```python
#Question 2: ##This is outputting two times, I dont know why :(
import numpy as np

#matrix input lene ka function
def InputPayoffs(row, col):
  PayoffMatrix = np.zeros((row, col), dtype=object)

  for i in range(row):
    for j in range(col):
      print(f"Enter payoffs at ({i}, {j}: p1 p2 )")
      p1, p2 = map(int, input().split())

      PayoffMatrix[i][j] = (p1, p2)
  print("\n")

  return PayoffMatrix

#strictly dominant strategy find karne ka function
def Dominant_strategy(matrix):
  rows = len(matrix)
  cols = len(matrix[0])

  p1_dominant = None
  p2_dominant = None


  #player 1 ka check
  for r1 in range(rows):
    dominant = True

    for r2 in range(rows):
      if r1 == r2:
        continue

      for c in range(cols):
        if matrix[r1][c][0] <= matrix[r2][c][0]:
          dominant = False
          break

      if not dominant:
        break

    if dominant:
      p1_dominant = r1



  #player 1 ka check
  for c1 in range(cols):
    dominant = True

    for c2 in range(cols):
      if c1 == c2:
        continue

      for r in range(rows):
        if matrix[r][c1][1] <= matrix[r][c2][1]:
          dominant = False
          break

      if not dominant:
        break

    if dominant:
      p2_dominant = c1


  if p1_dominant is not None:
    print(f"Player 1's dominant strategy is row {p1_dominant}")
  else:
    print("Player 1 has no dominant strategy")

  if p2_dominant is not None:
    print(f"Player 2's dominant strategy is column {p2_dominant}\n")
  else:
    print("Player 2 has no dominant strategy\n")




#Starting main :
# mat = InputPayoffs(2,2)
# Dominant_strategy(mat)

print("ADVERTISE VS NON-ADVERTISE:")
Dominant_strategy(advertise)
print("PRISONER'S DILEMA:")
Dominant_strategy(prisoners_dilema)
print("STAG HUNT:")
Dominant_strategy(stag_hunt)
```

```python
def nash_equilibrium(game):
  row = len(game)
  col = len(game[0])

  p1_best = set()
  p2_best = set()

  for c in range(col):
    max_payoff = max(game[r][c][0] for r in range(row))

    for r in range(row):
      if game[r][c][0] == max_payoff:
        p1_best.add((r,c))

  for r in range(row):
    max_payoff = max(game[r][c][1] for c in range(col))

    for c in range(col):
      if game[r][c][1] == max_payoff:
        p2_best.add((r,c))

  nash = p1_best & p2_best

  #to print coordinates where coords are presented as index and not whole number values
  '''if nash:
    print("Pure-Strategy Nash Equilibrium:")
    for coords in nash:
      print(coords)
  else:
    print("No Pure-Strategy Nash Equilibrium exists")'''


  if nash:
    print("Pure-Strategy Nash Equilibrium:")
    for (i,j) in nash:
      print(game[i][j])
  else:
    print("No Pure-Strategy Nash Equilibrium exists")


print("PRISONER'S DILEMA:")
nash_equilibrium(prisoners_dilema)
print("STAG HUNT:")
nash_equilibrium(stag_hunt)
print("CHICKEN:")
nash_equilibrium(chicken_game)
print("BATTLE OF SEXES:")
nash_equilibrium(sexes_battle)
print("A/B/C VS X/Y/X:")
nash_equilibrium(dosto)
```
